<h4>Question 1.1</h4>

On considère la fonction $f_{1} : \left(\omega_{0},\omega\right) \mapsto \frac{1}{n} \sum\limits_{i=1}^{n} \log\left(1+\exp\left(-y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right) + \frac{\rho}{2} \|\omega\|_{2}^{2}$ définie sur $\mathbb{R} \times \mathbb{R}^{p}$.

On cherche à calculer son gradient en $\left(\omega_{0},\omega\right) \in \mathbb{R} \times \mathbb{R}^{p}$.

Tout d'abord, on a clairement le fait que : $\frac{\partial ~ f_{1}}{\partial ~ \omega_{0}}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{-y_{i}\exp\left(-y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{1+\exp\left(-y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)} = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{-y_{i}}{1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}$.

Observons la fonction $f_{1}$ avec $\omega$ sous la forme $\left(\omega_{1},...\omega_{p}\right)$, vu que $\omega \in \mathbb{R}^{p}$, pour simplifier le calcul.

On peut ainsi écrire que : $f_{1}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \log\left(1+\exp\left(-y_{i}\left(\sum\limits_{k=1}^{p}x_{i,k} \omega_{k}+\omega_{0}\right)\right)\right) + \frac{\rho}{2} \sum\limits_{k=1}^{p} \omega_{k}^{2}$.

Par conséquent, $\forall k \in  [\![1;p]\!] $, $\left(\frac{\partial f_{1}}{\partial \omega}\left(\omega_{0},\omega\right)\right)_{k} = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{-y_{i}x_{i,k} \exp\left(-y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{1+\exp\left(-y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)} + \rho \omega_{k} = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{-y_{i}x_{i,k}}{1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)} + \rho \omega_{k}$.

On en déduit que $\frac{\partial ~ f_{1}}{\partial ~ \omega}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{-y_{i}x_{i}}{1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)} + \rho \omega$.

Finalement, on a donc $\nabla f_{1} \left(\omega_{0},\omega\right) = \left(\frac{1}{n} \sum\limits_{i=1}^{n} \frac{-y_{i}}{1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)},\frac{1}{n} \sum\limits_{i=1}^{n} \frac{-y_{i}x_{i}}{1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)} + \rho \omega\right)$.

Calculons maintenant sa hessienne en $\left(\omega_{0},\omega\right) \in \mathbb{R} \times \mathbb{R}^{p}$.

Tout d'abord, on a directement le fait que : $\frac{\partial^{2} f_{1}}{\partial \omega_{0}^{2}}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}}$.

De même, on a aussi : $\frac{\partial^{2} f_{1}}{\partial \omega \partial \omega_{0}}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} x_{i} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}}$ de manière évidente.

Ensuite, $\forall k \in  [\![1;p]\!] $, $\forall j \in  [\![1;p]\!] $, $\left(\frac{\partial^{2} f_{1}}{\partial \omega^{2}}\left(\omega_{0},\omega\right)\right)_{k,j} = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2}x_{i,k}x_{i,j} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}} + \delta\left(j,k\right) \rho$.

Autrement dit, on peut écrire que : $\frac{\partial^{2} f_{1}}{\partial \omega^{2}}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2}x_{i}x_{i}^{T} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}} + \rho 1\!\!1_{p}$.

Pour la dernière dérivée partielle, on voit que : $\forall k \in  [\![1;p]\!] $, $\left(\frac{\partial^{2} f_{1}}{\partial \omega_{0} \partial \omega}\left(\omega_{0},\omega\right)\right)_{k} = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} x_{i,k} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}}$.

Cela permet d'écrire que : $\frac{\partial^{2} f_{1}}{\partial \omega_{0} \partial \omega}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} x_{i} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}}$.

Finalement, on a donc $\nabla^{2}f_{1}\left(\omega_{0},\omega\right) = \begin{pmatrix}\frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}} & \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} x_{i}^{T} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}} \\ \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} x_{i} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}} & \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2}x_{i}x_{i}^{T} \exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)^{2}} + \rho 1\!\!1_{p}\end{pmatrix}$.

On remarque que si $p = 1$ et $n = 1$ alors $\det\left(\nabla^{2}f_{1}\left(\omega_{0},\omega\right)\right) = y^{2} \times \rho$. Cela signifie que si l'on considère $y$ comme non nul alors la matrice Hessienne de $f_{1}$ est symétrique positive.

Par rapport à la convexité de cette fonction, reportons-nous à la question 1.2 grâce à laquelle ce calcul s'en retrouve simplifié.

<h4>Suite de la question 1.1 sur la convexité de la fonction $f_{1}$ :</h4>

Soit maintenant un vecteur $A \in \mathbb{R}^{p+1}$ et un vecteur $\omega$ tel que décrit ci-dessus.

On a ainsi : $A^{T} \nabla^{2}f_{1}\left(\omega\right) A = A^{T} \left( \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2}x_{i}x_{i}^{T} \exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)\right)^{2}} + \begin{pmatrix} 0 & 0 \\ 0 & \rho 1\!\!1_{p}\end{pmatrix} \right) A = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} A^{T} x_{i}x_{i}^{T} A \exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)\right)^{2}} + A^{T} \begin{pmatrix} 0 & 0 \\ 0 & \rho 1\!\!1_{p}\end{pmatrix} A$.

En d'autres termes, $A^{T} \nabla^{2}f_{1}\left(\omega\right) A = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} \| x_{i}^{T} A \|_{2}^{2} \exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)\right)^{2}} + \rho \|A\|_{2}^{2} - \rho A_{0}^{2}$.

Cette formule est clairement positive, mais montrons qu'elle ne peut être nulle que si $A$ est le vecteur nul de $\mathbb{R}^{p+1}$.

Supposons par l'absurde que ce ne soit pas le cas. On a alors deux cas devant nous selon la valeur de $A_{0}$.

Si $A_{0} = 0$ alors $A^{T} \nabla^{2}f_{1}\left(\omega\right) A \geq \|A\|_{2}^{2} > 0$.

De même, si $A_{0} \neq 0$ alors $A^{T} \nabla^{2}f_{1}\left(\omega\right) A \geq \frac{1}{n} \sum\limits_{i=1}^{n} \frac{y_{i}^{2} A_{0}^{2} \exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)}{\left(1+\exp\left(y_{i}\left(x_{i}^{T}\omega\right)\right)\right)^{2}} > 0$.

Ceci étant absurde, nous avons donc prouvé la proposition ci-dessus.

On a ainsi montré que la matrice Hessienne de $f_{1}$ était bien symétrique définie positive.

On en conclut que la fonction $f_{1}$ est strictement convexe.

```python
CSV = dRU.load_diabetic_retinopathy("diabeticRetinopathy.csv")
X = np.insert(CSV[0],0,[1]*len(CSV[0]),axis=1)
Y = CSV[1]
n = len(X)
p = len(X[0])
rho = 1/n
```

```python
def objective(omega0,omega):
    return (1/n)*sum([log(1+exp(-Y[i]*((X[i].reshape(1,p)).dot(np.insert(omega,0,omega0))))) for i in range(n)]) + (rho/2) * np.linalg.norm(omega)**2
```

```python
def gradient(omega0,omega):
    return (1/n) * sum([- Y[i] * X[i] / (1+exp(Y[i]*((X[i].reshape(1,p)).dot(np.insert(omega,0,omega0))))) for i in range(n)]) + rho * (np.insert(omega,0,0))
```

```python
def hessienne(omega0,omega):
    specialEyeMatrix = rho * np.eye(p)
    specialEyeMatrix[0][0] = 0
    return (1/n) * sum([Y[i]**2 * (X[i].reshape(1,p).T).dot(X[i].reshape(1,p)) * exp(Y[i]*((X[i].reshape(1,p)).dot(np.insert(omega,0,omega0)))) / (1+exp(Y[i]*((X[i].reshape(1,p)).dot(np.insert(omega,0,omega0)))))**2 for i in range(n)]) + specialEyeMatrix
```

```python
f=lambda w:objective(w[0],w[1:p])
g=lambda w:gradient(w[0],w[1:p])
check_grad(f,g,[0.3 for k in range(p)])
```

<h4>Question 1.3</h4>

```python
def newton(initialValue):
    borne = 10**(-10)
    currentOmega0 = 1 * initialValue
    currentOmega = np.ones((p-1,1)) * initialValue
    currentGradient = gradient(currentOmega0,currentOmega)
    currentNormOfGradient = np.linalg.norm(currentGradient)
    listOfNorms = [currentNormOfGradient]
    while currentNormOfGradient >= borne:
        try:
            currentHessienne = hessienne(currentOmega0,currentOmega)
            currentHessienneInverted = np.linalg.inv(currentHessienne)
            hugeProduct = currentHessienneInverted.dot(currentGradient.T)
            currentOmega0 = currentOmega0 - hugeProduct[0]
            currentOmega = currentOmega - hugeProduct[1:].reshape(p-1,1)
            currentGradient = gradient(currentOmega0,currentOmega)
            currentNormOfGradient = np.linalg.norm(currentGradient)
            listOfNorms.append(currentNormOfGradient)
            print("Current norm of the gradient: %s." %(currentNormOfGradient),end="\r")
        except Exception as e:
            print("An error happened while processing the %sth iteration.\nError is as follows: %s." %(len(listOfNorms),e))
            break
    return currentOmega0,currentOmega,listOfNorms
```

```python
finalOmega0,finalOmega,listOfNorms = newton(0)
```

```python
print("We had to make %s iterations of the Newton algorithm to obtain a result." %(len(listOfNorms)))
```

```python
%matplotlib notebook
plt.plot([i for i in range(1,len(listOfNorms)+1)],listOfNorms)
plt.title("Evolution of the norm of the gradient with iterations of the Newton method")
plt.xlabel("Number of iterations")
plt.ylabel("Norm of the gradient (logarithmic scale)")
plt.yscale("log")
plt.show()
```

```python
def armijoLineSearch(currentOmega0,currentOmega,currentDirection,a,b,beta):
    currentValue = objective(currentOmega0,currentOmega)
    currentDirectionNorm = np.linalg.norm(currentDirection)**2
    s = b
    while True:
        xPlus0 = currentOmega0 - s * currentDirection[0]
        xPlus = currentOmega - s * currentDirection[1:].reshape(p-1,1)
        if objective(xPlus0,xPlus) > currentValue - s * beta * currentDirectionNorm:
            s = s * a
        else:
            return s

def newtonWithArmijo(initialValue):
    borne = 10**(-8)
    currentOmega0 = 1 * initialValue
    currentOmega = np.ones((p-1,1)) * initialValue
    currentGradient = gradient(currentOmega0,currentOmega)
    currentNormOfGradient = np.linalg.norm(currentGradient)
    listOfNorms = [currentNormOfGradient]
    s = 0.01
    while currentNormOfGradient >= borne:
        try:
            currentHessienne = hessienne(currentOmega0,currentOmega)
            currentHessienneInverted = np.linalg.inv(currentHessienne)
            hugeProduct = currentHessienneInverted.dot(currentGradient.T)
            s = armijoLineSearch(currentOmega0,currentOmega,currentGradient,0.5,2*s,0.5)
            currentOmega0 = currentOmega0 - s * hugeProduct[0]
            currentOmega = currentOmega - s * hugeProduct[1:].reshape(p-1,1)
            currentGradient = gradient(currentOmega0,currentOmega)
            currentNormOfGradient = np.linalg.norm(currentGradient)
            listOfNorms.append(currentNormOfGradient)
            print("Current norm of the gradient: %s." %(currentNormOfGradient),end="\r")
        except Exception as e:
            print("An error happened while processing the %sth iteration.\nError is as follows: %s." %(len(listOfNorms),e))
            break
    return currentOmega0,currentOmega,listOfNorms
```

<h4>Question 2.2</h4>

On peut écrire la fonction objectif ainsi : $F_{2} = f_{2} + g_{2}$ avec $f_{2}\left(\omega_{0},\omega\right) = \frac{1}{n} \sum\limits_{i=1}^{n} \log\left(1+\exp\left(-y_{i} \left(x_{i}^{T}\omega+\omega_{0}\right)\right)\right)$ et $g_{2}\left(\omega_{0},\omega\right) = \rho \|\omega\|_{1}$.

On peut alors calculer l'opérateur proximal de $g_{2}$ (i.e.) $\text{argmin}_{y\in\mathbb{R}^{n+1}} \left(\rho \|y\|_{1} + \frac{1}{2 \gamma} \|x - y\|_{2}^{2} \right)$ ce qui est faisable de manière analytique. 

Pour ce faire, on résout l'équation $\nabla \left( \rho \|y\|_{1} + \frac{1}{2 \gamma} \|x - y\|_{2}^{2} \right) = 0$.

On obtient le résultat suivant : $\text{prox}_{\gamma,g_{2}}\left(x\right) = \begin{pmatrix} ... \\ \left(\gamma p + x_{i} \right) \times \mathbb{1}_{\mathbb{R}_{-}^{\star}}\left(\gamma p + x_{i}\right) + \left( x_{i} - \gamma p \right) \times \mathbb{1}_{\mathbb{R}_{+}^{\star}}\left(x_{i} - \gamma p \right) \\ ... \end{pmatrix}$.

De plus, on montre que $\nabla f_{2}\left(\omega_{0},\omega_{1}, ~ ..., ~ \omega_{p}\right) = \nabla f_{2} \left( \omega\right) = - \frac{1}{n} \sum\limits_{k=1}^{n} x_{k} \frac{y_{k}}{1 + \exp\left(y_{k}x_{k}^{T}\omega\right)}$.

En outre, on montre que $\nabla^{2} f_{2} \left(\omega\right) = \frac{1}{n} X^{T} \text{Diag}\left(\alpha_{i}\right)_{1 \leq i \leq n} X$ avec $\forall i \in [\![1;n]\!]$, $\alpha_{i} = y_{i}^{2} \times \frac{\exp\left(- y_{i}x_{i}^{T}\omega\right)}{\left(1 + \exp\left(y_{i}x_{i}^{T}\omega\right)\right)^{2}} > 0$.

Puisque $\nabla^{2} g_{2} = 0$ et que $\nabla^{2} f_{2}$ est définie positive, on en déduit que la fonction objectif est convexe.

```python
import numpy as np
from time import time

def f2(x, y, w):
    n, p = X.shape
    if w.shape != (p,):
        print("pas la bonne taille")
        return
    
    tmp = np.array([exp(-y[i] * (X[i,:].T.dot(w))) for i in range(n)])
    
    val = np.mean(np.log(1 + tmp))
    return val
```

```python
def f2_grad(X, y, w):
    n, p = X.shape
    if w.shape != (p,):
        print("pas la bonne taille")
        return
    p = p-1
    
    tmp = np.array([exp(-y[i] * (X[i,:].T.dot(w))) for i in range(n)])
    tmp2 = -y * tmp/(1+tmp)
    grad =  np.mean(np.array([X[i,:]*tmp2[i] for i in range(n)]), axis=0)
    
    return grad
```

```python
def g2_prox_coef(x, gamma, rho):
    a = gamma*rho
    if x + a < 0:
        return x + a
    if x - a > 0:
        return x - a
    else:
        return 0
```

```python
def g2_prox(x, gamma, rho):
    return np.array([g2_prox_coef(xi, gamma, rho) for xi in x])
```

```python
def prox_search(f, f_grad, g_prox, x0, epsilon, a, b):
    x = x0.copy()
    lfunc, lgrad = f(x), f_grad(x)
    norm = np.linalg.norm(lgrad)
    n, gamma = 0, b
    e=1
    while e > epsilon and n < 500:
        xx = g_prox(x - gamma*lgrad, gamma)
        while f(xx) > lfunc + np.dot(lgrad.T, xx-x) + 1/(2*gamma)*(np.linalg.norm(x-xx)**2):
            gamma = gamma * a
            xx = g2_prox(x - gamma*lgrad, gamma, rho)
        
        e = np.linalg.norm(x-xx)
        
        if gamma != b :
            x = g2_prox(x - gamma/a*lgrad, gamma/a, rho)
        else:
            x = g2_prox(x - gamma*lgrad, gamma, rho)
            
        lfunc, lgrad = f(x), f_grad(x)
        gamma = b
        
        n+=1
        norm = np.linalg.norm(lgrad)
        
    return x, n
```

```python
rho = 0.1
w_init = 0.3*np.ones(p)
t = time()
w_min2, n = prox_search(lambda w : f2(X, Y, w), 
                     lambda w : f2_grad(X, Y, w), 
                     lambda x, gamma : g2_prox(x, gamma, rho), 
                     w_init, 
                     1e-10, 
                     0.5, 1)
print("On trouve le résultat en %d étapes soit %d secondes \n"%(n, time() - t), w_min2)
```
